# Text Augmentation: Random Word Swap

## What is Text Augmentation?

Text augmentation is a technique to create **new, slightly modified versions** of existing sentences.  
It helps **increase dataset size** and makes models more **robust and generalizable**.

---

## Simple Example: Random Word Swap

We will randomly swap two words in a sentence.

How it Works ?
<ul>
<li>Split the sentence into words.</li>

<li>Pick two random positions.</li>

<li>Swap the words at those positions.</li>

<li>Join the words back into a new sentence. </li></ul>


```python
import random

sentence = "The quick brown fox jumps over the lazy dog."
words = sentence.split()
i, j = random.sample(range(len(words)), 2)
words[i], words[j] = words[j], words[i]
augmented = " ".join(words)

print("Original:", sentence)
print("Augmented:", augmented)
```

# Text Augmentation: Synonym Replacement

## What is Synonym Replacement?

Synonym replacement is a technique where we **replace words in a sentence with their synonyms** to create a new sentence.  
It helps models **see more variety** without changing the meaning of the text.

### Example Concept

- Original: "The quick brown fox jumps over the lazy dog."  
- Possible Augmented: "The fast dark fox leaps over the idle dog."  

This keeps the **sentence meaning the same** but changes some words for variety.


```python
# Original sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Split sentence into words
words = sentence.split()

# Dictionary of synonyms
synonyms = {
    "quick": "fast",
    "lazy": "idle",
    "jumps": "leaps",
    "brown": "dark"
}

# Replace words with synonyms if they exist
for i, word in enumerate(words):
    # Remove punctuation for matching
    clean_word = word.strip(".!,?")
    if clean_word in synonyms:
        # Keep original punctuation
        punctuation = word[len(clean_word):]
        words[i] = synonyms[clean_word] + punctuation

# Join back into sentence
augmented = " ".join(words)

print("Original:", sentence)
print("Augmented:", augmented)
```

# Text Augmentation: Random Word Swap

## What is Random Word Swap?

Random word swap is a simple technique where we **swap two words randomly** in a sentence.  
It creates a **slightly different version** of the sentence to help models **learn better and generalize**.

### Example Concept

- Original: "The quick brown fox jumps over the lazy dog."  
- Possible Augmented: "The brown quick fox jumps over the lazy dog."  

This keeps the **meaning mostly the same**, but the **word order changes** slightly.


```python
import random

# Original sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Split sentence into words
words = sentence.split()

# Pick two random positions to swap
i, j = random.sample(range(len(words)), 2)

# Swap the words
words[i], words[j] = words[j], words[i]

# Join back into a sentence
augmented = " ".join(words)

print("Original:", sentence)
print("Augmented:", augmented)
```

# Text Augmentation: Random Word Swap (Multiple Swaps)

## What is Multiple Random Word Swap?

Instead of swapping just **one pair of words**, we can **swap multiple pairs** in a sentence to create **more variations**.  
This is useful when we want to **augment several sentences** in a dataset.

### Example Concept

- Original: "The quick brown fox jumps over the lazy dog."  
- Possible Augmented (2 swaps): "The brown quick fox over jumps the lazy dog."  

> More swaps = more variation, but still keeps the **meaning mostly intact**.


```python
import random

def random_swap(sentence, n_swaps=1):
    words = sentence.split()
    for _ in range(n_swaps):
        i, j = random.sample(range(len(words)), 2)
        words[i], words[j] = words[j], words[i]
    return " ".join(words)

# Test single sentence with multiple swaps
sentence = "The quick brown fox jumps over the lazy dog."
print("Random Swap:", random_swap(sentence, n_swaps=2))

# Apply to multiple sentences
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "She loves programming in Python.",
    "Machine learning is fun and powerful."
]

for s in sentences:
    print("Original:", s)
    print("Augmented:", random_swap(s))
    print()
```

# Text Augmentation: Back-Translation

## What is Back-Translation?

Back-translation is a technique where we:

1. Translate a sentence to another language.
2. Translate it back to the original language.

This often **paraphrases the sentence** while keeping the meaning intact.  
It is commonly used in **data augmentation for NLP**, especially when you want **more varied training sentences**.

### Example Concept

- Original: "The quick brown fox jumps over the lazy dog."  
- Back-translated Augmented: "The fast brown fox leaps over the lazy dog."  

Notice how some words are **replaced with synonyms** automatically.


```python
# Mock translation function
def translate(lang, text):
    if lang == 'fr':
        return "Le renard brun rapide saute par-dessus le chien paresseux."
    if lang == 'en':
        return "The fast brown fox leaps over the lazy dog."
    return text

# Original sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Step 1: Translate to French
fr_text = translate('fr', sentence)

# Step 2: Translate back to English
augmented = translate('en', fr_text)

print("Original:", sentence)
print("Back-translated Augmented:", augmented)
```

# Text Augmentation: Character-Level Typos

## What is Character-Level Augmentation?

Character-level augmentation introduces **small typos or character changes** in a sentence.  
This helps models become **robust to spelling mistakes** and **noisy text**, common in real-world data.

### Example Concept

- Original: "The quick brown fox."  
- Augmented: "The quock brown fox." (one letter changed randomly)


```python
import random

# Original sentence
sentence = "The quick brown fox."

# Convert sentence to list of characters
words = list(sentence)

# Pick a random index
i = random.randint(0, len(words)-1)

# Replace that character with a random letter
words[i] = random.choice('abcdefghijklmnopqrstuvwxyz')

# Join back into sentence
augmented = "".join(words)

print("Original:", sentence)
print("Augmented:", augmented)
```

# Text Augmentation: Paraphrasing

## What is Paraphrasing?

Paraphrasing is a technique where we **rewrite a sentence in a different way** while keeping the meaning the same.  
It helps models **see varied expressions** of the same idea, which improves **generalization**.

### Example Concept

- Original: "The quick brown fox jumps over the lazy dog."  
- Augmented: "A fast brown fox leaps over a sluggish dog."  

Notice how some words are replaced and the sentence structure may slightly change.


```python
# Mock paraphrase function
def paraphrase(text):
    return "A fast brown fox leaps over a sluggish dog."

# Original sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Apply paraphrasing
augmented = paraphrase(sentence)

print("Original:", sentence)
print("Augmented:", augmented)
```

# Text Augmentation: Embedding-Based Synonym Replacement

## What is Embedding-Based Replacement?

Embedding-based replacement uses **word embeddings (vectors)** to find **similar words** and replace them in a sentence.  
This allows models to see **semantically similar variations** of sentences, improving **generalization**.

### Example Concept

- Original: "The quick brown fox."  
- Augmented: "The fast brown fox."  

Here, "quick" is replaced by "fast" because their **embeddings are similar**.


```python
import torch

# Mock embeddings for words
embeddings = {
    "quick": torch.tensor([0.1, 0.2]),
    "fast": torch.tensor([0.11, 0.19])
}

# Original sentence
sentence = "The quick brown fox."

# Split sentence into words
words = sentence.split()

# Replace "quick" with "fast" if both exist in embeddings
for i, word in enumerate(words):
    if word in embeddings and "fast" in embeddings:
        words[i] = "fast"

# Join back into sentence
augmented = " ".join(words)

print("Original:", sentence)
print("Augmented:", augmented)
```

# Text Augmentation: Dataset-Level Random Word Swap

## What is Dataset-Level Augmentation?

We can apply augmentation to an **entire dataset** instead of just one sentence.  
This helps models **see more variations** and **generalize better**.  

In this example, we apply **random word swap** to each review in a dataset.

### Example Concept

- Original dataset:

| review       | label    |
|-------------|---------|
| Great movie! | positive|
| Bad plot.    | negative|

- Augmented dataset will contain **original + swapped sentences**.


```python
import pandas as pd
import random

# Original dataset
data = {
    'review': ['Great movie!', 'Bad plot.'],
    'label': ['positive', 'negative']
}
df = pd.DataFrame(data)

# Function to randomly swap words
def augment_text(text):
    words = text.split()
    if len(words) > 1 and random.random() > 0.5:
        i, j = random.sample(range(len(words)), 2)
        words[i], words[j] = words[j], words[i]
    return " ".join(words)

# Apply augmentation to dataset
augmented_rows = []
for _, row in df.iterrows():
    aug_text = augment_text(row['review'])
    augmented_rows.append({'review': aug_text, 'label': row['label']})

aug_df = pd.DataFrame(augmented_rows)

# Combine original + augmented
new_df = pd.concat([df, aug_df], ignore_index=True)
print(new_df)
```

# Text Augmentation: Rule-Based Entity Replacement

## What is Rule-Based Entity Replacement?

In some sentences, you have **specific entities** like names, locations, or dates.  
We can **replace these entities with other valid options** to create augmented sentences.  
This is especially useful for **Named Entity Recognition (NER)** or structured datasets.

### Example Concept

- Original: "Alice lives in Boston."  
- Augmented: "Alice resides in Paris."  

Notice how the **location is replaced** while the meaning remains similar.


```python
import random

def extract_and_augment(text):
    # Check if the pattern exists
    if "lives in" in text:
        parts = text.split(" lives in ")
        entity1 = parts[0]          # Extract person/entity
        entity2 = parts[1].strip('.')  # Extract location
        # Replace location with a random choice
        new_entity2 = random.choice(['New York', 'Paris'])
        return f"{entity1} resides in {new_entity2}."
    return text

# Original sentence
sentence = "Alice lives in Boston."

# Apply augmentation
augmented = extract_and_augment(sentence)

print("Original:", sentence)
print("Augmented:", augmented)
```
