# **What is NumPy?**

**NumPy = Numerical Python**

<ul><li>A Python library to handle numbers efficiently.</li>

<li>Why it exists: Python lists are slow for big math operations. NumPy arrays are fast, memory-efficient, and support multi-dimensional operations.</li></ul>

```python
import numpy as np

arr = np.array([1,2,3])
print(arr)
print(type(arr))
```

# What is an Array?

An array = a container of numbers, can be 1D, 2D, 3D, etc.

**Arrays = input features, images, sequences, labels. Operations = preprocessing, normalization, reshaping, feature scaling.**

```python
# 1D array → a row of numbers
a = np.array([1,2,3])
print("1D array:", a)
```

```python
# 2D array → table of numbers
b = np.array([[1,2],[3,4]])
print("2D array:\n", b)
```

```python
# 3D array → stack of tables
c = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print("3D array:\n", c)
```

# Numpy Operations


* Basic Math
* Element wise Operations
* Aggregate Operations
*  Reshaping
* Indexing & Slicing







```python
x = np.array([1,2,3,4])

# Basic math
print(x + 2)          # [3 4 5 6]
print(x * 3)          # [3 6 9 12]
```

```python

# Element-wise operations
y = np.array([10,20,30,40])
print(x + y)          # [11 22 33 44]
print(x * y)          # [10 40 90 160]

```

```python

# Aggregate operations
print(np.sum(x))      # 10
print(np.mean(x))     # 2.5
print(np.max(x))      # 4
print(np.min(x))      # 1
```

```python

# Reshape
z = np.arange(12)       # [0 1 2 ... 11]
print("\n",z)
print("\n")
print(z.reshape(3,4))   # 3 rows, 4 cols
```

```python
# Indexing & slicing
print(x[0])             # first element
print(x[1:3])           # slice [2 3]
```

# What is a Vector?

A vector is a 1-dimensional array.

Example: [23, 70, 5.8] → Age, weight, height of a person

```python
vector = np.array([23, 70, 5.8])
print("Vector:", vector)
print("Shape:", vector.shape)  # (3,)
```

# Intro to Pytorch

**What is a Tensor?**

A tensor = generalized array, can have any number of dimensions.

| Dimension | Example | Analogy | Shape |
|-----------|---------|---------|-------|
| 0D (scalar) | 5 | Single ball | () |
| 1D (vector) | [1,2,3] | Line of Lego blocks | (3,) |
| 2D (matrix) | [[1,2],[3,4]] | Tray of Lego blocks | (2,2) |
| 3D | [[[1,2],[3,4]], [[5,6],[7,8]]] | Stack of trays | (2,2,2) |
| 4D | Batch of 3D tensors | Multiple stacks | (batch, depth, rows, cols) |


*   All vectors and matrices are tensors, but tensors can be higher-dimensional.
*   In deep learning, every input/output is a tensor.



```python
import torch

# Single data point with 3 features → 1D tensor
x = torch.tensor([23,70,5.8])
print(x)
print(x.shape)
```

```python
# Batch of 4 data points → 2D tensor
batch_x = torch.tensor([
    [23,70,5.8],
    [45,80,5.5],
    [30,65,5.7],
    [25,72,5.6]
])

print(batch_x)
print(batch_x.shape)
```

```python
# Random tensor for CNN image input → 4D (batch, channels, height, width)
images = torch.rand((32,3,224,224))
print(images)
print(images.shape)
```

# Where do we use these?

| Tensor Type | Shape Example                           | Deep Learning Use  |
| ----------- | --------------------------------------- | ----------------- |
| 1D vector   | `(features,)`                           | Single data point |
| 2D matrix   | `(batch_size, features)`                | Input to MLP      |
| 3D tensor   | `(sequence, features, embedding)`       | RNN input         |
| 4D tensor   | `(batch_size, channels, height, width)` | CNN input (images)|
| 5D tensor   | `(batch_size, frames, channels, H, W)`  | Video input       |
